```python
def get_raklama_data(campaign_ids, begin: str, end: str):
    headers = {
        "Authorization": TOKEN
    }
    
    total_res = []

    start = datetime.fromisoformat(begin)
    finish = datetime.fromisoformat(end)
    current = start
    while True:
        nxt = min(current + timedelta(days=30), finish)  # +30 → интервал 31 день
        begin_date, end_date = current.strftime("%Y-%m-%d"), nxt.strftime("%Y-%m-%d")
        current = nxt + timedelta(days=1)
        print(current)
        campaign_ids_start = 0
        campaign_ids_end = min(50, len(campaign_ids))
        while True:
            print("campaign_ids_window = ", campaign_ids_start, campaign_ids_end)
            params = {
                "ids":       ",".join([str(company['advertId']) for company in campaign_ids[campaign_ids_start:campaign_ids_end]]),   
                "beginDate": begin_date,
                "endDate":   end_date
            }
            url = "https://advert-api.wildberries.ru/adv/v3/fullstats"

            response = requests.get(url, headers=headers, params=params, timeout=20)
            response.raise_for_status()                 
            res = response.json()     
            total_res += res
            if campaign_ids_end == len(campaign_ids):
                break
            campaign_ids_start += 50
            campaign_ids_end = min(campaign_ids_start+50, len(campaign_ids))
            sleep(22)
        if current > finish:
            break
        sleep(22)
    return total_res
    
DF = get_raklama_data(campaign_ids, "2025-10-13", "2025-12-01")
```
